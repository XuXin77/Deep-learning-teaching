## 对数组的的分批读入

```python
def generator_array(image_data, mask_data, batch_size):
    samples_num = image_data.shape[0]
    index = list(range(samples_num))

    random.seed(1)
    random.shuffle(index)
    random.shuffle(index)
    
 
    for i in range((samples_num // batch_size + 1)):
        image_arr = []
        mask_arr = []
        for j in range(batch_size):
            try:
                item = index[(i+1)*(j+1)-1] 
                # img=np.resize(img,(256,256))
                # img = np.reshape(img,img.shape + (1,))
                img = image_data[item,:,:,:]
                # print(item,item.replace(image_path,mask_path))
                mask = mask_data[item,:,:]
                # mask=np.resize(mask,(256,256))
                # mask = np.reshape(mask,img.shape  + (1,))
                img = img / 255
                mask = mask /255
                mask[mask > 0.5] = 1
                mask[mask <= 0.5] = 0
                image_arr.append(img)
                mask_arr.append(mask)
            except IndexError:
                i = 0
                item = index[(i+1)*(j+1)-1] 
                img = image_data[item,:,:,:]
                # img=np.resize(img,(256,256))
                # img = np.reshape(img,img.shape + (1,))
                # print(item,item.replace(image_path,mask_path))
                mask = mask_data[item,:,:]
                # mask=np.resize(mask,(256,256))
                # mask = np.reshape(mask,img.shape  + (1,))
                img = img / 255
                mask = mask /255
                mask[mask > 0.5] = 1
                mask[mask <= 0.5] = 0
                image_arr.append(img)
                mask_arr.append(mask)

        image_arr = np.array(image_arr)
        mask_arr = np.array(mask_arr)
        yield image_arr, mask_arr 
```

* 产生数组样本 images_data， masks_data

```python
def Load_Color_Data(image_path, mask_path):
    image_name_arr = glob.glob(os.path.join(image_path,"*.jpg"))
    image_arr = []
    mask_arr = []
    
    for index,item in enumerate(image_name_arr):
        img = io.imread(item)
        # img=np.resize(img,(256,256))
        # img = np.reshape(img,img.shape + (1,))
        # print(item,item.replace(image_path,mask_path))
        mask = io.imread(item.replace(image_path,mask_path))
        # mask=np.resize(mask,(256,256))
        # mask = np.reshape(mask,img.shape  + (1,))
        img = img / 255
        mask = mask /255
        mask[mask > 0.5] = 1
        mask[mask <= 0.5] = 0
        image_arr.append(img)
        mask_arr.append(mask)

    image_arr = np.array(image_arr)
    mask_arr = np.array(mask_arr)
    return image_arr, mask_arr
```
